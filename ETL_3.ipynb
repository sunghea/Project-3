## Part 3: Explore

```python
import re
# Specify the file path
file_name = "insurance_raw_data.json"
file_path = os.path.join(directory, file_name)

# Load the JSON data from the file
with open(file_path, "r", encoding="utf-8") as file:
    data = json.load(file)

# Define the keywords to search for
keywords = [
    "Professional and hobby associations",
    "Insurance agency",
    "Insurance company",
    "Home insurance agency",
    "Financial planner",
    "Insurance broker",
    "Corporate office"
]

# Define the regular expression pattern to extract phone numbers
phone_number_pattern = r"\(\d{3}\) \d{3}-\d{4}"

# Initialize a list to store the filtered data
filtered_data = []

# Iterate over each item in the data list
for item in data:
    # Check if "div_2" exists in the item
    if "div_2" in item:
        # Check if any of the keywords exist in "div_2" value
        for keyword in keywords:
            if keyword in item["div_2"]:
                # If found, add the item to the filtered data list
                item["div_4"] = keyword
                filtered_data.append(item)
                break  # Stop searching for keywords once found
    
    # Check if "div_3" exists in the item
    if "div_3" in item:
        # Extract phone number from "div_3" value using regular expression
        phone_number_match = re.search(phone_number_pattern, item["div_3"])
        if phone_number_match:
            phone_number = phone_number_match.group()
            # Add the phone number to "div_5"
            item["div_5"] = phone_number

            
# Specify the file path
file_name = "filtered_insurance_raw_data.json"
file_path = os.path.join(directory, file_name)

# Write the filtered JSON data to the file
with open(file_path, "w", encoding="utf-8") as file:
    json.dump(filtered_data, file)
print("JSON data has been successfully saved to:", file_path)
```
